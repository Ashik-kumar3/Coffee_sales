```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('employe.csv',sep=';')
df.head()
```

```python
df['Education'].head()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,10))

axs[0,0].hist(df['Education'],color='y')
axs[0,0].set_title('employe')
axs[0,0].set_xlabel('Education')
axs[0,0].set_ylabel('Count')

axs[0,1].hist(df['City'],color='g')
axs[0,1].set_title('City')
axs[0,1].set_xlabel('City')
axs[0,1].set_ylabel('Count')

axs[1,0].hist(df['Age'],bins=5,color='r',rwidth=0.5)
axs[1,0].set_title('Age')
axs[1,0].set_xlabel('Education')
axs[1,0].set_ylabel('Count')

axs[1,1].hist(df['Gender'],color='k')
axs[1,1].set_title('Gender')
axs[1,1].set_xlabel('Gender')
axs[1,1].set_ylabel('Count')
```
