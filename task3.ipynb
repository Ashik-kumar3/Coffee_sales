```python
import pandas as pd
df=pd.read_csv('netflix_titles.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['release_year'],patch_artist=True,showmeans=True)
plt.title('Money')
plt.show()
```

```python
qrtl1=df['release_year'].quantile(0.25)
qrtl3=df['release_year'].quantile(0.75)
```

```python
IQR=qrtl3-qrtl1
IQR
```

```python
lw = max(qrtl1 - 1.5*IQR,df['release_year'].min())
lw
```

```python
uw = min(qrtl3 + 1.5*IQR,df['release_year'].max())
uw
```

```python
out = df[(df['release_year']<lw) | (df['release_year']>uw)]
out.head()
```

```python
df1 = df[(df['release_year']>lw) & (df['release_year']<uw)]
df1.head()
```
