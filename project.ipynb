OBJECTIVE :- To do an analysis on the E-commerce sales data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df=pd.read_csv('Sales Data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
df['City']=df['City'].str.strip()
```

```python
df['City'].unique()
```

```python
def City(row):
    if row == 'New York City':
        return 0
    elif row == 'San Francisco':
        return 1
    elif row == 'Atlanta':
        return 2
    elif row == 'Portland':
        return 3
    elif row == 'Dallas':
        return 4
    elif row == 'Los Angeles':
        return 5
    elif row == "Boston":
        return 6
    elif row == 'Austin':
        return 7
    elif row == 'Seattle':
        return 8
  
```

```python
df['City']=df['City'].apply(City)
```

```python
df['City'].unique()
```

```python
df.head()
```

```python
df['Product Category'].unique()
```

```python
df['Product Category']=df['Product Category'].replace({'Laptops and Computers' :0,'Home Appliances':1, 'Charging Cables':2, 'Monitors':3,'Batterie':4,'Audio Devices':5,
                                                      'Phones and Accessories':6,  'Entertainment Devices':7})
```

```python
df['Product Category']=df['Product Category'].astype(int)
```

```python
df['Product Category'].unique()
```

```python
df['Time of Day'].unique()
```

```python
df['Time of Day']=df['Time of Day'].replace({'Morning':0, 'Afternoon':1,'Evening':2,'Night':3})
```

```python
df['Time of Day'] = df['Time of Day'].astype(int)
```

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
monthly_sales=df.groupby('Month')['Sales'].sum()
monthly_sales.index = monthly_sales.index.astype(str)
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(monthly_sales,color='k')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month',)
plt.ylabel('Revenue')
plt.xticks(ticks=range(0,12), labels=months, rotation=45)
plt.grid(color='r')
plt.show()
```

Inference: The plot indicates that revenue rises gradually until April, followed by a decline from May to September. After September, revenue recovers and reaches its highest level in December.

```python
sns.scatterplot(data=df, x='Quantity Ordered', y='Sales')
plt.title("Quantity vs Sales")
plt.show()
```

```python
top = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(10)

top.plot(kind='bar',color='r')
plt.title("Top 10 Selling Products")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
hourly_sales = df.groupby('Hour')['Sales'].sum().reset_index()

sns.lineplot(data=hourly_sales, x='Hour', y='Sales')
plt.title("Sales by Hour")
plt.show()
```

```python
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()
```

```python
Time=df.groupby('Time of Day')['Sales'].sum()
Time.index = Time.index.astype(str)
plt.bar(Time.index,Time,width=0.3,color='y')
plt.title('Timely Revenue Trend')
plt.xticks(Time.index,['morning','afternoon','evening','night'])
plt.xlabel('Time of Day')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
Time
```

Inference: The timely revenue analysis shows that revenue peaks keeps increasing from morning to night. The maximum peak is at night.

```python
pivot = df.pivot_table(values='Sales', index='Hour', columns='Month', aggfunc='sum')

sns.heatmap(pivot,cmap='coolwarm')
plt.title("Sales Heatmap (Hour vs Month)")
plt.show()
```

```python
product_sales = df.groupby('Product Category')['Quantity Ordered'].sum()
plt.figure(figsize=(7,7))
plt.pie(product_sales.values,
        labels= [ 'Laptops and Computers',        'Home Appliances',        'Charging Cables',
               'Monitors',               'Batterie',          'Audio Devices',
 'Phones and Accessories',  'Entertainment Devices'],
        autopct='%1.1f%%',   
        startangle=90,
        pctdistance=0.8)
plt.title('Quantity Ordered by Product Category')
plt.show()
```

Inference:- This piechart showing that the batterie is one of the product category which the people ordered the most with around 28%. The Charging cables and hold same percentage around 23.
