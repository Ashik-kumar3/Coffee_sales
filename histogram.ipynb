```python
import matplotlib.pyplot as plt
```

```python
age=[10,12,23,12,31,22,34,54,32,12,33,24,15,16,14]
plt.hist(age)
plt.xlabel("Age",color='red')
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

```python
plt.hist(age,rwidth=.5,color='red',bins=[10,14,20,25,30,35,40])
plt.xlabel("Age",color='red')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('employe.csv',sep=';')
df.head()
```

```python
age_freq=df[(df['City']=='Bangalore')&(df['Education']=='Bachelors')].groupby('Age').count()['City']
age_freq
```

```python
plt.hist(age_freq,rwidth=.5,color='y',bins=[10,40,80,150,250,300,350,400])
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```
