```python
import pandas as pd
df=pd.read_csv('employe.csv',sep=';')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df = df[(df['City'] == 'Bangalore') & (df['JoiningYear'] == 2017)]

male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']
```

```python
male_count_bachelors = len(male[male['Education'] == 'Bachelors'])
male_count_masters= len(male[male['Education'] == 'Masters'])
male_count_phd = len(male[male['Education'] == 'PHD'])
```

```python
female_count_bachelors = len(female[female['Education'] == 'Bachelors'])
female_count_masters= len(female[female['Education'] == 'Masters'])
female_count_phd = len(female[female['Education'] == 'PHD'])
```

```python
total_male = ()
total_male+=(male_count_bachelors,male_count_masters,male_count_phd)
total_female = ()
total_female+=(female_count_bachelors,female_count_masters,female_count_phd)
```

```python

x1 = [0, 1, 2]
x2 = [0.1, 1.1, 2.1]

plt.bar(x1,total_male,label='Male',color='b',width=.1)
plt.bar(x2,total_female,label='Female',color='y',width=.1)
plt.legend()
plt.xticks(x1,['Bachelors','Masters','PHD'])
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Education and Count')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('employe.csv',sep=';')
df.head()
```

```python
count_female=df[(df['City']=='Bangalore') & (df['JoiningYear']==2017) & (df['Gender']=='Female')].groupby('Education').count()['Gender']
```

```python
count_male=df[(df['City']=='Bangalore')&(df['JoiningYear']==2017) & (df['Gender']=='Male')].groupby('Education').count()['Gender']
```

```python
count_male.index
```

```python
import numpy as np
x=np.array([])
for i in range(len(count_male.index)):
    x=np.append(x,i)
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.bar(x,count_male,width=.2,color='r',label='Male')
plt.bar(x+0.2,count_female,width=0.2,color='k',label='Female')
plt.legend()
plt.xticks(x,count_male.index)
plt.xlabel('Education')
plt.ylabel('Count')
plt.show()
```
